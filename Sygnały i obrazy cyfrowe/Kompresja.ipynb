```python
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.image import imread
from skimage import color
```

```python
image = plt.imread("kot.jpg")
plt.rcParams['figure.figsize'] = [5, 5]
plt.rcParams.update({'font.size': 10})
plt.figure()
plt.imshow(image)
plt.axis('off')
```

```python
grey = color.rgb2gray(image)
plt.figure()
plt.imshow(grey, cmap='gray')
plt.axis('off')
```

```python
grey_fourier = np.fft.fft2(grey)
grey_fourier_sort = np.sort(np.abs(grey_fourier))

for keep in (0.1, 0.05, 0.01, 0.002):
    thresh = grey_fourier_sort[int(np.floor((1-keep)*len(grey_fourier_sort)))]
    ind = np.abs(grey_fourier)>thresh
    low = grey_fourier * ind
    Alow = np.fft.ifft2(low).real #compresed image
    plt.figure()
    plt.imshow(Alow, cmap='gray')
    plt.axis('off')
    plt.title('Compressed Image: keep = ' + str(keep * 100) + '%')
```
